# Outlier Detection

```python
# Create boxplots for numerical columns

plt.figure(figsize=(15,5))

plt.subplot(1,3,1)
sns.boxplot(y=df["Sales"])
plt.title("Sales")

plt.subplot(1,3,2)
sns.boxplot(y=df["Profit"])
plt.title("Profit")

plt.subplot(1,3,3)
sns.boxplot(y=df["Quantity"])
plt.title("Quantity")

plt.tight_layout()
plt.show()
```

## Outlier Analysis

- Boxplots were used to inspect numerical variables for outliers.
- Sales contains several high-value transactions, which are expected in retail data.
- Profit includes both high profits and significant losses due to discounts and business operations.
- Quantity shows only a few high-value observations.
- Since these outliers represent genuine business transactions rather than data entry errors, no outliers were removed.

# Sales Analysis


```python
# Total Sales by Category

category_sales = (
    df.groupby("Category")["Sales"]
      .sum()
      .sort_values(ascending=False)
)

plt.figure(figsize=(8,5))

ax = sns.barplot(
    x=category_sales.index,
    y=category_sales.values,
)

# Add value labels on top of bars
for i, value in enumerate(category_sales.values):
    ax.text(i, value + 10000, f"{value:,.0f}", ha='center', fontsize=10)

plt.title("Total Sales by Category", fontsize=14, fontweight="bold")
plt.xlabel("Category")
plt.ylabel("Total Sales")

plt.tight_layout()

plt.savefig("../Images/sales_by_category.png", dpi=300)

plt.show()
```

### Insight

- Technology generated the highest total sales among all product categories.
- Furniture ranked second in terms of revenue.
- Office Supplies generated the lowest sales.
- This suggests that Technology products contribute the most to the company's revenue.

# Profit Analysis

```python
# Profit by Region

region_profit = (
    df.groupby("Region")["Profit"]
      .sum()
      .sort_values(ascending=False)
)

plt.figure(figsize=(8,5))

ax = sns.barplot(
    x=region_profit.index,
    y=region_profit.values,
)

# Add value labels
for i, value in enumerate(region_profit.values):
    ax.text(i, value + 1000, f"{value:,.0f}", ha='center', fontsize=10)

plt.title("Total Profit by Region", fontsize=14, fontweight="bold")
plt.xlabel("Region")
plt.ylabel("Total Profit")

plt.tight_layout()

plt.savefig("../Images/profit_by_region.png", dpi=300)

plt.show()

print(region_profit)
```

### Insight

- The West region generated the highest total profit.
- The East region also contributed significantly to overall profitability.
- The Central region recorded the lowest profit among all regions.
- The company should investigate factors affecting profitability in the Central region, such as discounts, product mix, or operational costs.

```python
segment_sales = (
    df.groupby("Segment")["Sales"]
      .sum()
      .sort_values(ascending=False)
)

plt.figure(figsize=(8,5))

ax = sns.barplot(
    x=segment_sales.index,
    y=segment_sales.values,
)

for i, value in enumerate(segment_sales.values):
    ax.text(i, value + 10000, f"{value:,.0f}", ha="center", fontsize=10)

plt.title("Total Sales by Customer Segment", fontsize=14, fontweight="bold")
plt.xlabel("Customer Segment")
plt.ylabel("Total Sales")

plt.tight_layout()

plt.savefig("../Images/sales_by_segment.png", dpi=300)

plt.show()

print(segment_sales)
```

### Insight

- The Consumer segment generated the highest sales among all customer segments.
- Corporate customers contributed the second-highest revenue.
- Home Office customers generated the lowest sales.
- The company should continue focusing on the Consumer segment while exploring strategies to increase sales in the Home Office segment.

# Monthly Sales Trend

```python
# Create Year-Month column
df["Year_Month"] = df["Order Date"].dt.to_period("M").astype(str)

monthly_sales = (
    df.groupby("Year_Month")["Sales"]
      .sum()
)

plt.figure(figsize=(14,6))

sns.lineplot(
    x=monthly_sales.index,
    y=monthly_sales.values,
    marker="o",
    linewidth=2.5
)

plt.title("Monthly Sales Trend", fontsize=15, fontweight="bold")
plt.xlabel("Year-Month")
plt.ylabel("Total Sales")

plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.5)

plt.tight_layout()

plt.savefig("../Images/monthly_sales_trend.png", dpi=300)

plt.show()
```

### Insight

- Monthly sales show an overall increasing trend from 2014 to 2017.
- Several seasonal peaks are visible, especially during the final months of each year.
- The highest sales were recorded in late 2017.
- These patterns indicate strong seasonal demand, suggesting that inventory and marketing campaigns should be planned around peak sales periods.

# Profit by Category

```python
# Profit by Category

category_profit = (
    df.groupby("Category")["Profit"]
      .sum()
      .sort_values(ascending=False)
)

plt.figure(figsize=(8,5))

ax = sns.barplot(
    x=category_profit.index,
    y=category_profit.values
)

for i, value in enumerate(category_profit.values):
    ax.text(i, value + 2000, f"{value:,.0f}", ha="center", fontsize=10)

plt.title("Total Profit by Category", fontsize=14, fontweight="bold")
plt.xlabel("Category")
plt.ylabel("Total Profit")

plt.grid(axis="y", linestyle="--", alpha=0.5)

plt.tight_layout()

plt.savefig("../Images/profit_by_category.png", dpi=300)

plt.show()

print(category_profit)
```

### Insight

- Technology is the most profitable product category.
- Office Supplies also contributes significantly to overall profit.
- Furniture generates high sales but comparatively low profit, indicating lower profit margins or higher discounting.

```python
top_states = (
    df.groupby("State")["Profit"]
      .sum()
      .sort_values(ascending=False)
      .head(10)
)

plt.figure(figsize=(10,6))

ax = sns.barplot(
    x=top_states.values,
    y=top_states.index
)

plt.title("Top 10 States by Profit", fontsize=14, fontweight="bold")
plt.xlabel("Profit")
plt.ylabel("State")

plt.grid(axis="x", linestyle="--", alpha=0.5)

plt.tight_layout()

plt.savefig("../Images/top_states_profit.png", dpi=300)

plt.show()
```

### Insight

- California and New York are the company's strongest profit-generating states.
- The top three states contribute a significant share of total profit.
- Marketing and expansion strategies in these regions could further increase profitability.

```python
top_products = (
    df.groupby("Product Name")["Sales"]
      .sum()
      .sort_values(ascending=False)
      .head(10)
)

plt.figure(figsize=(12,6))

ax = sns.barplot(
    x=top_products.values,
    y=top_products.index
)

plt.title("Top 10 Products by Sales", fontsize=14, fontweight="bold")
plt.xlabel("Sales")
plt.ylabel("Product")

plt.grid(axis="x", linestyle="--", alpha=0.5)

plt.tight_layout()

plt.savefig("../Images/top_products_sales.png", dpi=300)

plt.show()
```

### Insight

- The Canon imageCLASS 2200 Advanced Copier generated the highest sales.
- High-end office equipment dominates the top-selling products.
- These products should receive focused inventory and promotional planning.

```python
plt.figure(figsize=(8,6))

sns.heatmap(
    df[["Sales","Profit","Quantity","Discount"]].corr(),
    annot=True,
    cmap="coolwarm",
    linewidths=0.5
)

plt.title("Correlation Heatmap")

plt.tight_layout()

plt.savefig("../Images/correlation_heatmap.png", dpi=300)

plt.show()
```

### Insight

- Sales and Profit show a moderate positive correlation.
- Discounts negatively impact profit, indicating that excessive discounting reduces profitability.
- Quantity has only a weak relationship with sales and profit.
